This assignment explores Cooling Centers around LA County. We pull in data from [LA Heatwaves Data Hub](https://la-heatwaves-gisucla.hub.arcgis.com/) which designates cooling centers intended for extreme heat events. The data comes in GeoJson format and we will create a map of the centers around LA

```python
import pandas as pd 
import geopandas as gpd
```

Read into the GeoJson file for Formal County Community Centers 

```python
countyCenters = gpd.read_file("data/Formal_County_Centers.geojson") 
```

Read into the GeoJson file for Formal County Libraries. Libraries are different than community centers but both act as effective places for cooling during extreme heat event

```python
countyLibraries = gpd.read_file("data/Formal_County_Library_Footprint.geojson")
```

```python
countyCenters.info()
```

The data provide information on the Center's name, lat and long location points, geometry, and FID.

```python
countyCenters.cntr_name
```

```python
countyLibraries.info()
```

The data for the county Libraries provides much more information, including building  height, builing area, etc. However, these columns are less important because we really only need the location.

Now we will join the dataframes together. We will use the concat method in pandas to concatenate or stack the dataframes together to create one large geopandas data frame. There will only be a few columns that overlap, which are the only ones we need to focus on.

```python
joined = pd.concat([countyCenters, countyLibraries])
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
joined.info()
```

```python
joined.FID
```

By using .info(), we are able to see how many null and non-null values are in each column after joining the dataframes together. The cntr_name, FID, match_addr, x, y, and geometry columns each have 95 values, which tells us that these are the only columns which both datasets match on ny column name aka the only shared data types that were collected. 

```python
joined.cntr_name
```

Quick check to make sure the data can be plotted

```python
joined.plot()
```

Now we want to map the locations using folium. This will help us to visualize where the data are. 

```python
import folium 
```

```python
m = folium.Map(location=[34.2,-118.2], 
               zoom_start = 9,
               tiles='CartoDB positron', 
               attribution='CartoDB')

# plot chorpleth over the base map
folium.Choropleth(
                  geo_data=joined,
                  fill_color='BuPu',
                  line_weight=0.1, 
                  fill_opacity=0.8,
                  line_opacity=0.2, # line opacity (of the border)
                  legend_name='Cooling Centers and Libraries in LA County').add_to(m)    # name on the legend color bar

folium.LayerControl(position = 'topright', collapsed=True, autoZIndex=True).add_to(m) #need capital T for true or capital F for False

m
```

The map helps us to see that most of the centers are clustered in East LA, which makes sense given that heat is more extreme inland LA.
